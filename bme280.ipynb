```python
import pandas as pd #for import csv
```

```python
bme_data = pd.read_csv("data1_4_night.csv") #before 'upload files'
print(bme_data)
```

```python
import matplotlib.pyplot as plt
```

```python
x = bme_data['d4']
y_temp = bme_data['d1']
y_humi = bme_data['d2']
y_press = bme_data['d3']
```

```python
fig, ax1 = plt.subplots()
p1 = ax1.plot(x,y_temp,label='pop',marker="o",color="blue")
ax1.set_title("WeatherData_1/4_night")
ax1.set_xlabel('min')
ax1.set_ylabel('C°')
ax2 = ax1.twinx()  # 2つのプロットを関連付ける
ax3 = ax1.twinx()  # 2つのプロットを関連付ける
rspine = ax3.spines['right'] # 3軸対応
rspine.set_position(('axes', 1.2)) # y軸の位置をずらす
ax3.get_yaxis().get_major_formatter().set_useOffset(False) # 気圧の数字を省略形から本来の形へ 
p2 = ax2.plot(x,y_humi,label='pop',marker="o",color="red")
ax2.set_ylabel('%')
p3 = ax3.plot(x,y_press,label='pop',marker="o",color="green")
ax3.set_ylabel('Pa')
ax1.legend((p1[0], p2[0],p3[0]), ("Temp", "Humi","Press"), loc=2)
```
